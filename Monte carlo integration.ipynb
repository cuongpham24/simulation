```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Set the random seed for reproducibility
np.random.seed(24)
```

# Monte Carlo integration
Approximate the integral of the function $f(x)=\sin(\pi x)$ over $[0, 1]$ by randomly choosing n points $x_i$ uniformly distributed over the interval.

Similar to Riemann sums, we devide the area under the curve to n rectangulars with the same width.
$$w=\frac{1-0}{n}$$
Monte Carlo integration can be used to approximate this integral as follows:
$$ \int_{0}^{1}\sin (\pi x)dx=\frac{1}{n} \sum_{x_i=0}^n \sin(\pi x_i) $$

As the number of random points increases, the approximation of the integral converges to the true value.


```python
# Define f(x)
def f(x):
    return np.sin(np.pi * x)

# Integrating over [0, 1]
def integration(n):
    # Generate uniformly distributed n random points
    x = np.random.uniform(0, 1, size=n)
    y = f(x)
    # Monte Carlo integration
    integral_approximation = (1 / n) * np.sum(y)

    return x, y, integral_approximation
```

```python
### Test case
n = 10000
x, y, integral_approximation = integration(n)

print(f"With n = {n}, the esimated integral is {integral_approximation}")

### Visulize
# Draw f(x)
x_0 = np.linspace(0, 1, n)
y_0 = f(x_0)

# Select random 100 points for visulization
sample_size = 50
i = np.random.choice(n, sample_size, replace=False)

# Graph f(x) and random x_i
plt.figure(figsize=(7, 4))
plt.plot(x_0, y_0, color="red")
plt.scatter(x[i], y[i], color="blue", alpha=0.5)
plt.vlines(x[i], 0, y[i], linestyles="dashed", alpha=0.5)

plt.title(f'Monte Carlo Integration\nIntegral Approximation: {np.round(integral_approximation, 4)}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

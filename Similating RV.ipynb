```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2024)
```

# Discrete Random Variable

```python
n = 10000
U = np.random.uniform(0, 1, size=n)
V = np.random.uniform(0, 1, size=n)
Z = np.sqrt(-2 * np.log(U)) * np.cos(2 * np.pi * V)
```

```python
plt.hist(Z, bins=50, edgecolor='black');
```
